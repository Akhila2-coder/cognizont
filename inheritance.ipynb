```python
class parent:
    def fun1(self):
        print("this function is in parents class")
class child (parent):
    def fun2(self):
        print("this function is in child class")
object=child()
object.fun1()
object.fun2()
```

```python
class parent:
    def func1(self):
        print("this function belongs to parents class")
class child(parent):
    def func1(self):
        print("this function belongs to child class")
object=child()
object.func1()
```
